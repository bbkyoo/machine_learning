#### Sneakers

```python
import requests
from bs4 import BeautifulSoup

url = 'https://search.musinsa.com/category/018002'

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
res = requests.get(url, headers=headers)
res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml')

sneakers_nums = soup.find_all('li', {'class':'li_box'})
sneakers_names = soup.find_all('p', {'class':'list_info'})
sneakers_prices = soup.find_all('p', {'class':'price'})
sneakers_comments = soup.find_all('span', {'class':'count'})


sneakers_num_list = []
sneakers_name_list = []
sneakers_ori_price_list = []
sneakers_sale_price_list = []
sneakers_comment_list = []

for i in range(10):
    sneakers_num = int(sneakers_nums[i]['data-no'])
    sneakers_name = sneakers_names[i].find('a', {'name':'goods_link'})['title']
    
    if sneakers_prices[i].find('del'):
        sneakers_ori_price = sneakers_prices[i].find('del').get_text()
        sneakers_sale_price = sneakers_prices[i].get_text().split()[1]
        sneakers_sale_price = int(sneakers_sale_price.replace(',', '')[:-1])
        
    else:
        sneakers_ori_price = sneakers_prices[i].get_text().strip()
        sneakers_sale_price = None
        
    sneakers_ori_price = int(sneakers_ori_price.replace(',', '')[:-1])
        
    sneakers_comment = int(sneakers_comments[i].get_text().strip().replace(',', ''))

    sneakers_num_list.append(sneakers_num)
    sneakers_name_list.append(sneakers_name)
    sneakers_ori_price_list.append(sneakers_ori_price)
    sneakers_sale_price_list.append(sneakers_sale_price)
    sneakers_comment_list.append(sneakers_comment)
    

import pandas as pd

sneakers = pd.DataFrame({'p_num':sneakers_num_list, 'p_name':sneakers_name_list, 'ori_price':sneakers_ori_price_list,
                         'sale_price':sneakers_sale_price_list, 'comments':sneakers_comment_list})
sneakers
```
